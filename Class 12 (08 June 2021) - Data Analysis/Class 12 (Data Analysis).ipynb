### **Crosstab: In Python, a crosstab is a tabulation of two different categorical variables.**

### **Ploting with seaborn**

If we write normalize = TRUE, it will give us the percentage values.

```python
import matplotlib.pyplot as plt
import seaborn as sns

pd.crosstab(df['international plan'], df['churn'])

sns.countplot(x = 'international plan', hue = 'churn', data = df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pd.crosstab(df['international plan'], df['churn'], normalize=True)

sns.countplot(x = 'international plan', hue='churn', data= df)
```

### **Crosstab using and (&), or (|)**

```python
df['international plan']=df['international plan'].replace('yes', 1)
df['international plan']=df['international plan'].replace('no', 0)
df['voice mail plan']=df['voice mail plan'].replace('yes', 1)
df['voice mail plan']=df['voice mail plan'].replace('no', 0)

pd.crosstab(df['international plan'] | df['voice mail plan'], df['churn'])
```

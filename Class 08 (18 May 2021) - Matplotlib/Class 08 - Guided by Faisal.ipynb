```python
import numpy as np
import matplotlib.pyplot as plt
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y,color='green')
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y,marker='o')
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y,marker='.')
plt.show()
```

```python
from matplotlib.pyplot import figure
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
figure(figsize=(10,5),dpi=100)
plt.plot(x,y,marker='.')
plt.show()
```

Subplot

```python
fig,axs=plt.subplots(2)

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

fig.suptitle("hsdgshadga")

axs[0].plot(x,y)
axs[1].plot(x,-y)
```

```python
fig,axs=plt.subplots(2,2)

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
figure(figsize=(10,5),dpi=100)
fig.suptitle("hsdgshadga")

axs[0][0].plot(x,y)
axs[0][1].plot(x,-y)
axs[1][0].plot(x,y)
axs[1][1].plot(x,-y)
```

```python
fig,(axs1,axs2)=plt.subplots(1,2)

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
fig.suptitle("hsdgshadga")
fig.tight_layout(pad=3.0)
axs1.plot(x,y)
axs2.plot(x,-y)
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y,marker='.',color='g')
plt.plot(x,-y,marker='.',color='b')
plt.legend(['green','blue'],loc="upper right")
plt.show()
```

margins

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)

plt.plot(x,y,marker='.',color='g')
plt.plot(x,-y,marker='.',color='b')
plt.legend(['green','blue'],loc="upper right")
plt.margins(x=0.5,y=0.2)
plt.show()
```

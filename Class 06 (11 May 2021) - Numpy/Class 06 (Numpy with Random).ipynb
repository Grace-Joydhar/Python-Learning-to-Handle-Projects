**What is a Random Number?**
Random number does NOT mean a different number every time. Random means something that can not be predicted logically.

```python
import numpy as np
x = np.random.random((3,4))
print(x)
```

**Integers:** The **randint() method** takes a size parameter where you can specify the shape of an array.

```python
import numpy as np
x = np.random.randint(100, size = (5))
print(x)
```

```python
import numpy as np
x = np.random.randint(100, size = (3,4))
print(x)
```

**Floats:** **The rand() method** also allows you to specify the shape of the array.

```python
import numpy as np
x = np.random.rand(5)
print(x)
```

```python
import numpy as np
x = np.random.rand(3,4)
print(x)
```

**Generate Random Number From Array**:
The choice() method takes an array as a parameter and randomly returns one of the values.

The choice() method also allows you to return an array of values.
Add a size parameter to specify the shape of the array.

```python
import numpy as np
x = np.random.choice([1,2,3,4,5], size = (3,5))
print(x)
```

**numpy.zeros**(shape, dtype=float, order='C', *, like=None): 
Return a new array of given shape and type, filled with zeros.

```python
import numpy as np
x = np.zeros((3,4))
print(x)
```

**numpy.ones**(shape, dtype=float, order='C', *, like=None): 
Return a new array of given shape and type, filled with ones.

```python
import numpy as np
x = np.ones((3,4))
print(x)
```

**np.empty():** Return a new array of given shape and type, without initializing entries.

```python
import numpy as np
x = np.empty([2,3], dtype='int')
print(x)
```

```python
import numpy as np
x = np.array([1,2,3,4])
print(x.cumprod())
# intermediate results 1, 1*2
# total product 1*2*3 = 6 , 1*2*3*4 = 24
```

```python
import numpy as np
x = np.array([1,2,3,4])
print(x.cumsum())
# intermediate results 1, 1+2 = 3
# total product 3+3 = 6 , 6+4 = 10
```

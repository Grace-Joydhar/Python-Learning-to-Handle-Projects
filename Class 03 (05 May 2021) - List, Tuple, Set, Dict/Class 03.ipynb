Functions for List are:

len(): T0 find out the length of the list

append()	Adds an element at the end of the list

insert()	Adds an element at the specified position

count()	Returns the number of elements with the specified value

extend()	Add the elements of a list (or any iterable), to the end of the current list

index()	Returns the index of the first element with the specified value

pop()	Removes the element at the specified position

remove()	Removes the item with the specified value

sort()	Sorts the list

reverse()	Reverses the order of the list

copy()	Returns a copy of the list

clear()	Removes all the elements from the list
del() Deletes a list

slice() Cut a list values from a point.

```python
#len(): T0 find out the length of the list
z = [1,2,3, "BD", (1, "Beautiful", 2, "Nature"), 4,5]
len(z)
```

```python
#append()	Adds an element at the end of the list
a = [1, "python", 2, "learn"]
a.append("properly")
print(a)
```

```python
#insert() adds a element by indicating index number
b = [1, "python", 2, "learn"]
b.insert(4, 3)
b.insert(5, "easily")
print(b)
```

```python
#count(): Returns the number of elements with the specified value
c = [1,2,3,1,1,1,1]
print(c.count(1))
```

```python
#extend(): To add extra string or list with the existing list.
#1st way to do extend:
d = [1, "learn"]
d.extend([2, "python"])
print(d)
```

```python
#2nd way to do extend
e = [1, "learn"]
f = [2, "python"]
e.extend(f)
print(e)
```

```python
#3rd way to do extend
g = [1, "learn"]
g += [2, "python"]
print(g)
```

```python
#index(): Returns the index number of the specified value
h = ["learn", "python", (1,2,3)]
print(h.index("python"))
```

```python
#find a value for a specific index
i = ["learn", "python", (1,2,3)]
print(i[0])
print(i[2])
print(i[2][1])
```

```python
#pop(): Only pop() will remove the last value. but suppose pop(1) will remove the value which is present in no. 1 index
j = [1,2,3,4,5]
print(j.pop()) #remove the last value
print(j)

print(j.pop(2)) #remove value from index 2
print(j)
```

```python
#del(): Also remove values from index
x = [1,2,3,4,5,6]
del x[2] #will delete the value from 2nd index
print(x)
```

```python
#remove(): Removes the item with the specified value
k = [1,2,3,(4,5,6)] #Will remove 3
k.remove(3)
print(k)
```

```python
#sort(): will sort the values from lower to upper values
l = [10,30,20,60,40,50]
l.sort()
print(l)
```

```python
#sort in reverse
m = [10,30,20,60,40,50]
m.reverse() #Sometimes does not print the right answer
print(m)

n = [10,30,50,20,60,10]
n.sort(reverse=True) #Best way to reverse any list items.
print(n)
```

```python
#clear(): removes all the values from the list. but list will exit in the memory
o = [1,2,3, (4,5,6,7,8)]
o.clear()
print(o)
```

```python
#del() wll delete the list from memory. so it will show error.
p = [1,2,3,4]
del p
print(p)
```

```python
#slice(): Cut a list values from a point.
s = [1,2,3, True, False, "python", "learn", 4,5]
print(s[1:5])
print(s[:6])
print(s[4:])
print(s[-1])
print(s[-6: -2])
```

**Dictionary:**

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and does not allow duplicates.

```python
mydict = {'name' : 'XYZ', 'age' : 23} #name and age are key, XYZ and 23 are keys.
print(mydict)
```

```python
#Does not contain duplicate values
mydict = {'name' : 'XYZ', 'age' : 29, 'age' : 24} #will remove 1st key and value
print(mydict)
```

**Functions of Dictionary:**
1. get(): Also used for finding a value of a key
2. keys(): To find all the keys present in the dictionary
3. values(): To find all the values present in the dictionary
4. update(): to change any value
5. pop(): to remove a key and value together
6. popitem(): to remove a key and value together from the end of the dict
7. clear(): to clear the dict
8. del() to remove any key : value or to totally delete any dict
9. copy(): to copy a dict

```python
#To find a value of a key:
a = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
print(a['name'])
```

```python
#get(): Also used for finding a value of a key
b = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
print(b.get('age'))
```

```python
#keys(): To find all the keys present in the dictionary
c = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
print(c.keys())
```

```python
#values(): To find all the values present in the dictionary
c = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
print(c.values())
```

```python
#Add a new item
d = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
d['DOM'] = ['August']
print(d)
```

```python
#change a item
#Normal way
e = {'name' : 'XYZ', 'age' : 29, 'Phone' : '01713'}
e['age'] = 24
print(e)
```

```python
#change a item
#update() Function: to change a value
f = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
f.update({'phone' : '01879'})
print(f)
```

```python
#pop(): to remove a key and value together
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
g.pop('name')
print(g)
```

```python
#popitem(): to remove a key from the last of the dict
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
g.popitem()
print(g)
```

```python
#clear(): to clear the dict
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
g.clear()
print(g)
```

```python
#del() to remove any key : value or to totally delete any dict
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
del g['name']
print(g)
```

```python
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
del g
print(g)
```

```python
#copy(): to copy a dict
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
n = g.copy()
print(g)
print(n)
```

```python
#dict(): copy a full list with dict() function
g = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
o = dict(g)
print(g)
print(o)
```

**Loop in Dictionary**
1. items(): To print all the keys and values together by using items() function

```python
#No 1 way: To print all the keys
keys = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
for i in keys:
  print(i)
```

```python
#No 2 way: To print all the keys
k = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
for i in k.keys():
  print(i)
```

```python
#No 1 way: To print all the values
values = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
for i in values:
  print(values[i])
```

```python
#No 2 way: To print all the values
v = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
for i in v.values():
  print(i)
```

```python
#items(): To print all the keys and values together by using items() function
m = {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}
for i, j in m.items():
  print(i,':', j)
```

```python
#Nested Dict:
p = {"Child1" : {'name' : 'XYZ', 'age' : 29, 'phone' : '01713'}, 
      "Child2" : {'name' : 'ABC', 'age' : 20, 'phone' : '01879'}, 
      "Child3" : {'name' : 'DEF', 'age' : 34, 'phone' : '01914'}}
print(p)
```

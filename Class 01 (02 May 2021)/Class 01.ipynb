To avoid space before printing anything, we canuse %

```python
a = "Four"
print("Two and two makes: %s"%a)
```

Add two String

```python
a = "Hello " + "World"
print(a)
```

```python
a = int(input("Enter any number: "))
print(a)
```

User Input (More than one variable)

```python
a , b = input("Enter any number: ").split()
print(a,b)
```

```python
print("Bangladesh is a", end=" ")
print("beautiful country")
```

# **Pandas Series**
A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.

```python
import pandas as pd

a = [1,4,8]

myvar = pd.Series(a)
print(myvar)

# First column is the index number column (called labels), generate automatically. 
```

# **Key/Value Objects as Series**
You can also use a key/value object, like a dictionary, when creating a Series.

```python
import pandas as pd

calories = {'Banana': '23gm', 'Apple': '20gm', 'Sweet': '50gm'} #keys will work as index number

myvar = pd.Series(calories)
print(myvar)
print("Only the calories in banana is:",myvar['Banana'])
```

# **What is a DataFrame?**
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
import pandas as pd

data = {
        'Fruits' : ['mango', 'apple', 'banana'],
        'calories' : [20, 40, 10],
        'color' : ['green', 'red', 'yellow']
        }
mylabels = ['f1', 'f2', 'f3']

myvar = pd.DataFrame(data, mylabels)
print(myvar)
```

**Locate Row**: Pandas use the loc attribute to return one or more specified row(s)

```python
import pandas as pd

data = {
        'Fruits' : ['mango', 'apple', 'banana'],
        'calories' : [20, 40, 10],
        'color' : ['green', 'red', 'yellow']
        }
mylabels = ['f1', 'f2', 'f3']

myvar = pd.DataFrame(data, mylabels)
print(myvar.loc['f1'])  #If we locate only one index, then we need to use one [] bracket. This will return a dataframe.
print("\n")
print(myvar.loc[['f2', 'f3']]) ##If we locate more than one index, then we need to use two [[]] brackets. This will return a series
```

**Replace Using Mean, Median, or Mode** Calculate the Mean, Median or Mode and replace any empty values with it.
1. Mean = the average value (the sum of all values divided by number of values).
2. Median = the value in the middle, after you have sorted all values ascending.
3. Mode = the value that appears most frequently.

```python
#Fill all the empty calls with 0.9
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/iris.csv')

x = df['petal_width'].mean() #Same for median and mode
df['petal_width'].fillna(x, inplace=True)
print(df)
```

# **Pandas - Fixing Wrong Data**
If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.

It doesn't have to be wrong, but taking in consideration that this is the data set of someone's workout sessions, we conclude with the fact that this person did not work out in 450 minutes.

**Replacing Values**: Suitable for small dataset

```python
#Read a file from drive
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/dirtydata.csv')
df.loc[7, 'Duration'] = 45
print(df)
```

**Replacing Values with**: Suitable for large dataset

```python
#Read a file from drive
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/dirtydata.csv')
for elements in df.index:
  if df.loc[elements, 'Duration'] > 60:
    df.loc[elements, 'Duration'] = 60
print(df)
```

**Removing Rows**

```python
#Read a file from drive
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/Colab_Notebooks/Sample_data/dirtydata.csv')
for elements in df.index:
  if df.loc[elements, 'Duration'] > 120:
    df.drop(elements, inplace= True)
print(df)
```

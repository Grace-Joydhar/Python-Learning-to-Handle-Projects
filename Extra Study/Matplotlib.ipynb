**What is Matplotlib?** :
Matplotlib is a low level graph plotting library in python that serves as a visualization utility. 

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

```python
import matplotlib
print(matplotlib.__version__)
```

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

**Matplotlib Plotting**:
The x-axis is the horizontal axis. The y-axis is the vertical axis.

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])

plt.plot(x, y)
plt.show()
```

**Multiple Points**
You can plot as many points as you like, just make sure you have the same number of points in both axis.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,4,6,8,9])
y = np.array([10, 45, 60, 20, 90])

plt.xlabel("X Axis")  #give a name for x axis
plt.ylabel("Y Axis") #give a name for y axis
plt.title("X-Y Graph") #name of the graph

plt.plot(x,y)
plt.show()
```

# **Figure**
With the figure size we can define the graph sizes in inches 

figure(figsize = (2,2)), First 2 is length in inches and second 2 is width in inches.

figure(figsize = (2,2), dpi = 100) , dpi means dot per inches). this is mention to specify the dot numbers i per inches.

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import math

x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

figure(figsize = (7,3), dpi = 100)

plt.suptitle("Figure and Fig size Study")
plt.plot(x,y)
plt.show()
```

# **[Matplotlib Markers](https://)**
1. **Marker:** You can use the keyword argument ***marker*** to emphasize each point with a specified marker(circle, star, point, +, | ...)
2. **Line Style (ls):** how will be the graph line looks like (ls ='-'/ '--'/ '..'/'-.')
3. **color (c):** color means the line color or marker color(r, w, k = blacek, g, b=blue, m, y)
You can also usse Hexadecimal color values (ex: #4CAF50), [140 supported color names.](https://) (hotpink)

4. **Line width or lw:** To mention the width of the line

5. ***Format Strings fmt: ***
You can also use the shortcut string notation parameter to specify the marker This parameter is also called fmt, and is written with this syntax:

--> ***marker|line|color*** example: (o:r) here o is circle, : is line design and r is color red.

6. **Marker size(ms)**
7. **Marker edge color (mec)**
8. **Marker face color (mfc)**

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([1,3,5,7])
y = np.array([10,30, 50, 70])

plt.plot(x,y, marker = 'o') #marker design
plt.show()
plt.plot(x,y, marker ='+') 
plt.show()
```

# **Matplotlib Labels and Title**
1. **Label:** With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis
2. **fontdict parameter** in xlabel(), ylabel(), and title() to set font properties for the title and labels. *Ex: font = {'family':'Arial', 'size': 20, 'color':'green'}*
3. loc parameter: in title() to position the title. that means ustify the title (center, left, right). Default is center

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,4,6,8,9])
y = np.array([10, 45, 60, 20, 90])

font1 = {'family':'Arial', 'size': 20, 'color':'green'}
font2 = {'family':'Arial', 'size': 15, 'color' : 'blue'}

plt.title("X-Y Graph", fontdict= font1, loc = 'right') #name of the graph
plt.xlabel("X Axis", fontdict= font2)  #give a name for x axis
plt.ylabel("Y Axis", fontdict= font2) #give a name for y axis

plt.plot(x,y)
plt.show()
```

# **Matplotlib Subplots:**
Display Multiple Plots
With the subplots() function you can draw multiple plots in one figure.

The subplots() function takes one arguments that describes the layout of the figure. 

fig,axs=plt.subplots(2), there will be 2 subplots, fig will measure the fig number, axs will define the positions

axs[0].plot(x,y)
axs[1].plot(x,y) , in this way we can define the axs.

**Distance between graphs**: use fig/plt.tight_layout(pad = 1.08)

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig
import numpy as np 

fig,axs = plt.subplots(2)

x = np.array([10,30,20,50])
y = np.array([2,5,10,20])

figure(figsize=(10,5), dpi= 100)
fig.suptitle = "Subplots Study"

axs[0].plot(x,y)
axs[1].plot(x,-y)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig

fig,(axs1, axs2) = plt.subplots(1,2)

x = np.array([10,30,20,40])
y = np.array([20,40,60,80])

figure(figsize=(5,5),dpi=100)
fig.suptitle("Graph alongside another graph")
fig.tight_layout(pad= 2.00)

axs1.plot(x,y, color ='g')
axs2.plot(x,-y, color = 'r')

fig.legend(['x,y', 'x,-y'], loc='best')
plt.show()
```

You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure as fig
import numpy as np

fig, axs = plt.subplots(2,2)

x = np.array([10,20,50,40])
y = np.array([40,50,20,30])

fig.suptitle("multiple Subplots")

#figure(figsize = (10,10), dpi = 10)

axs[0][0].plot(x,y)
axs[0][1].plot(x,-y)
axs[1][0].plot(y,x)
axs[1][1].plot(y,-x)

plt.show()
```

# **Matplotlib Pie Charts**
**Creating Pie Charts**: use the pie() function to draw pie charts.
By default the plotting of the first wedge starts from the x-axis and move counterclockwise

Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: x/sum(x)

**1. labels:** name of the parts.

**2. startangle:** From where the chart will be started (0,90,180,270)

**3. explode:** the distence of the part from the center.

**4. shadow:** create a shadow for every part of the chart by using shadows perameter to True

**5. colors:** Change the color with colors perameter

**6. labeldistance:** Distance from pie chart tothe names of differet parts

**7. legend:** To add a list of explanation for each wedge, use the **legend()** function

**8. title:** to add a title of the legend header.

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.array([10,12,40,30])
my_labels = ['Apple', 'banana', 'jackfruit', 'cherry']

plt.pie(y, labels = my_labels)
plt.show()
```

**Start Angle:**
The startangle parameter is defined with an angle in degrees, default angle is 0. 

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.array([10,12,40,30])
my_labels = ['Apple', 'banana', 'jackfruit', 'cherry']

plt.pie(y, labels = my_labels, startangle=90)
plt.show()
```

**Explode**: 
The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.array([10,30,20,40])
my_labels = ['apple', 'banana', 'jackfruit', 'cherry']
my_explode = [0, 0.2, 0, 0]

plt.pie(y, labels = my_labels, explode = my_explode)
plt.show()
```

**Shadow:** create a shadow for every parte of the perameter. For this, make shadow to True.

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.array([10,30,20,40])
my_labels = ['apple', 'banana', 'jackfruit', 'cherry']
my_explode = [0, 0.2, 0, 0]

plt.pie(y, labels = my_labels, explode = my_explode, shadow=True)
plt.show()
```

**Colors:** Set different colors to every single part by using ***colors perameter.***

```python
import numpy as np
from matplotlib import pyplot as plt

y = np.array([10,30,20,40])
my_color = ['#7FFFD4', '#5F9EA0', '#556B2F', '#20B2AA']
my_labels = ['apple', 'banana', 'jackfruit', 'cherry']
my_explode = [0, 0.2, 0, 0]

plt.pie(y, labels = my_labels, explode = my_explode, shadow=True, colors = my_color)
plt.show()
```

**Legend**
To add a list of explanation for each wedge, use the ***legend() function:***

**location:** loc() is used to specfy the place where the legend will be shown.

```python
import numpy as np
from matplotlib import pyplot as plt
a = np.array([10,40,30,50])
my_color = ['#7FFFD4', '#5F9EA0', '#556B2F', '#20B2AA']
my_labels = ['apple', 'banana', 'jackfruit', 'cherry']
my_explode = [0, 0.2, 0, 0]

plt.pie(a,colors = my_color, labels=my_labels, labeldistance= 1.1,  explode = my_explode, shadow = True) 
plt.legend(title = 'Four Fruits')
plt.show()
```

# **ggplot**
It is used to create a background in the back of graph. we can use different background design such as ggplot, classic, fivethirtyeight, bmh, dark_background, grayscale and so on.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig

fig,(axs1, axs2) = plt.subplots(1,2)

x = np.array([10,30,20,40])
y = np.array([20,40,60,80])

plt.style.use('ggplot')
#plt.style.use('classic')

figure(figsize=(5,5),dpi=100)
fig.suptitle("Graph alongside another graph")
fig.tight_layout(pad= 2.00)

axs1.plot(x,y, color ='g')
axs2.plot(x,-y, color = 'r')

fig.legend(['x,y', 'x,-y'], loc='best')
plt.show()
```

# **Margins**

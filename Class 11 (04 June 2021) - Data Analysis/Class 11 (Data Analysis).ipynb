### **QUES: How much time (on average) do churned users spend on the phone during daytime?**

```python
# for i in df['churn']:
#   if i == True:
#     print(df['total day minutes'].mean())

avgTime= df[df['churn'] == True]['total day minutes'].mean()
print("Avg time is: ", avgTime)
```

### **QUES: What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?**

```python
maxLength = df[(df['churn'] == False) & (df['international plan'] == 'no')]['total intl minutes'].max()
print("maximum length is:",maxLength)
```

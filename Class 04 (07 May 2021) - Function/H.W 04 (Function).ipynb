2. Write a C program to find diameter, circumference and area of circle using functions.

```python
import math

r = float(input("Enter a radius value: "))

def diameter(r):
  return r * 2 #Rule: 2*radius

def circumference(r):
  return math.pi * diameter #Rule: Diameter * pi

def circle(r):
  return math.pi * r**2 #Rule: pi * r^2

diameter = diameter(r)
circumference = circumference(r)
circle = circle(r)
  
print(diameter)
print(circumference)
print(circle)
```

3. Write a C program to find maximum and minimum between two numbers using functions.

```python
#normal way
def max_num(a,b):
    if a>b:
        return a
    else:
        return b

def min_num(a,b):
    if a<b:
        return a
    else:
        return b

max_result = max_num(20,10)
min_result = min_num(20,10)

print("Max is:", max_result, "Min is:", min_result)
```

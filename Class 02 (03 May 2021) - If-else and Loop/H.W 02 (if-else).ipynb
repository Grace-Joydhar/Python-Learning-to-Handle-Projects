4.	Write a C program to check whether a number is divisible by 5 and 11 or not.

```python
num = int(input("Enter your number: "))
print(num)

if (num%5 == 0) and (num%11 != 0):
    print("Divisible by 5 but not divisible by 11")
elif (num%5 != 0) and (num%11 == 0):
    print("Not divisible by 5 but divisible by 11")
elif (num%5 == 0) and (num%11 == 0):
    print("Divisible by 5 and 11")
elif (num%5 != 0) and (num%11 != 0):
    print("Not divisible by 5 and 11")
else:
    print("Wrong Information")
```

6.	Write a C program to check whether a year is leap year or not.

```python
#If a year is exactly divisible by 4 and if a year is ended up with 00 at last, and possible to divide by 400, then it's a leap year.

year = int(input("Enter a year: "))

if year%4 == 0 or year%400 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```

14.	Write a C program to check whether the triangle is equilateral, isosceles or scalene triangle.

```python
a = int(input("Enter 1st angle: "))
b = int(input("Enter 2nd angle: "))
c = int(input("Enter 3rd angle: "))

total = (a +b +c)
print("Total is: ", total)

if total == 180:
    if a == b ==c:
        print("Equilateral Triangle")
    elif a == b or b == c or c == a:
        print("Isosceles Triangle")
    elif a != b !=c:
        print("Scalene Triangle")
else:
    print("Wrong Information. Triangle value should be 180 Degree")
```

16.	Write a C program to input basic salary of an employee and calculate its Gross salary according to following:

Basic Salary <= 10000 : HRA = 20%, DA = 80%
Basic Salary <= 20000 : HRA = 25%, DA = 90%
Basic Salary > 20000 : HRA = 30%, DA = 95%

```python
basic_salary = float(input("Enter your basic salary: "))

if basic_salary <= 10000:
    HRA = basic_salary * (20 /100)
    DA = basic_salary * (80 /100)
    total_salary = (basic_salary + HRA + DA)
    print("Your Gross Salary is: ",total_salary)
elif basic_salary <= 20000:
    HRA = basic_salary * (25 /100)
    DA = basic_salary * (90 /100)
    total_salary = (basic_salary + HRA + DA)
    print("Your Gross Salary is: ",total_salary)
elif basic_salary > 20000:
    HRA = basic_salary * (30 /100)
    DA = basic_salary * (95 /100)
    total_salary = (basic_salary + HRA + DA)
    print("Your Gross Salary is: ",total_salary)
else:
    print("Wrong Input")
```

17.	Write a C program to input electricity unit charges and calculate total electricity bill according to the given condition:
For first 50 units Rs. 0.50/unit
For next 100 units Rs. 0.75/unit
For next 100 units Rs. 1.20/unit
For unit above 250 Rs. 1.50/unit
An additional surcharge of 20% is added to the bill

```python
x = float(input("Enter you charge: "))

if x<=50:
    result = x * 0.50
    charge = (result * 20) / 100
    Total_bill = result + charge
    print("Total bill is: ",Total_bill)
elif x>50 and x<=150:
    result = x * 0.75
    charge = (result * 20) / 100
    Total_bill = result + charge
    print("Total bill is: ",Total_bill)
elif x>150 and x<=250:
    result = x * 1.20
    charge = (result * 20) / 100
    Total_bill = result + charge
    print("Total bill is: ", Total_bill)
elif x>250:
    result = x * 1.50
    charge = (result * 20) / 100
    Total_bill = result + charge
    print("Total bill is: ", Total_bill)
else:
    print("Error occured")
```

1. If-else problem:

```python
x = 2
if x == 2:
  print("2")
elif x == 3:
  print("3")
else:
  print("None")
```

2. abs() Function: Only print positive values. If a variable has negative value, then also, it will print positive value.

```python
x = -2.53
print(abs(x))
```

3. round() Function: It will give only round value. Suppose, if you insert 2.50, it will be 2. But if you insert 2.51, it will print 3.

```python
x = 2.51
print(round(x))
```

4. Ceiling and floor are the functions of math library. So we need to import math first.

Ceilng: It will give you the nearest upper value. If 2.5, then will print 3.

Floor: It will give you the nearest lower value. If 2.5, will print 2

```python
import math
x = 2.50
result = math.ceil(x)
print(result)
```

```python
import math
x = 2.5
result = math.floor(x)
print(result)
```

5. **Some Other Functions for String:**

isalpha() to check if a string contains only
characters and is not empty

isalnum() to check if a string contains
characters or digits and is not empty

isdecimal() to check if a string contains digits
and is not empty

lower() to get a lowercase version of a string

islower() to check if a string is lowercase

upper() to get an uppercase version of a string

isupper() to check if a string is uppercase

title() to get a capitalized version of a string

startsswith() to check if the string starts with a
specific substring

endswith() to check if the string ends with a
specific substring

replace() to replace a part of a string

split() to split a string on a specific character
separator

strip() to trim the whitespace from a string

join() to append new letters to a string

find() to find the position of a substring

```python
a = "abc"
print(a.isalpha())
```

```python
b = "123"
print(b.isdigit())
print(b.isdecimal())
```

```python
c = "a1b2c3"
print(c.isalnum())
```

```python
#Upper: lower to upper
d = "abc"
print(d.upper())
```

```python
#isupper: check all letters are upper or not. True or False
e = "ABC"
print(e.isupper())
```

```python
#Lower: upper to lower
f = "ABC"
print(f.lower())
```

```python
#islower: check all letters are lower or not. True or False
g = "abc"
print(g.islower())
```

```python
#title: To get upper case version of the first letter only among a sentence.
h = "python is fun"
print(h.title())
```

```python
#startswith(): to check if the string starts with a specific substring or not. True or False
i = "Bangladesh is a small country"
print(i.startswith("Bangladesh"))
```

```python
#endswith(): to check if the string ends with a specific substring. True or False.
j = "Bangladesh is a small country"
print(j.endswith("country"))
```

```python
#replace(): replace a word with another word
k = "Python is fun"
print(k.replace("fun", "awesome"))
```

```python
#split(): Split convert a string into a list, Separate characters.
l = "Python is fun"
print(l.split())
```

```python
#strip(): Remove the whitespaces between string
m = "     Sky is blue"
print(m.strip())
```

```python
#join(): Add a letter into a string.
n = ("Rivers", "are", "nice")
print("+".join(n))
o = "Sky is blue"
print("=".join(o))
```

```python
#find(): To find out the position/index number from where a string starts.
p = "Birds are humming"
print(p.find("humming"))
```

6. For Loop:

```python
#Print 1 to 10.
for i in range(1,11):
  print(i)
  i = i+1
```

```python
for i in range(1,10):
  for j in range(1,i):
    print(i,j)
```

7. While Loop:

```python
#print 1 to 10.
i = 1
while i <= 10:
  print(i)
  i = i+1
```

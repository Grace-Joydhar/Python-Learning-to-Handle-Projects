13. Write a C program to swap first and last digits of a number.

```python
num = int(input("Enter a num: "))

first = num
last = num

while first >= 10:
  first = first//10
print("First number is:",first)

if last >= 0:
  last = last % 10
print("Last number is:",last)

first, last = last, first

print("\nAfter swaping:")
print("First number is:",first)
print("Last number is:",last)
```

17. Write a C program to check whether a number is palindrome or not.

```python
num = int(input("Enter a number: "))
rev = 0
temp = num

while num > 0:
  last = num%10
  rev = (rev*10) + last
  num = num//10
if temp == rev:
  print("Palindrome Number")
else:
  print("Not a palindrome number")
```

21. Write a C program to find HCF (GCD) of two numbers. - Highest common factor

```python
num1= int(input("Enter a number: "))
num2= int(input("Enter a number: "))

hcf = 0

#we first determine the smaller of the two numbers since the H.C.F can only be less than or equal to the smallest number. 

if num1 < num2:
  min = num1
else:
  min = num2

#Run a for loop to find the divisible numbers and store the number in hcf. 
for i in range(1, min):
  if (num1%i == 0) and (num2%i == 0):
    hcf = i
    print(i, end = " ")
  i = i + 1
print("\nHcf of the number is: ",hcf)
```

22. Write a C program to find LCM of two numbers. - Least common multiple

```python
num1= int(input("Enter a number: "))
num2= int(input("Enter a number: "))

hcf = 0

#we first determine the smaller of the two numbers since the H.C.F can only be less than or equal to the smallest number. 

if num1 < num2:
  min = num1
else:
  min = num2

#Run a for loop to find the divisible numbers and store the number in hcf. 
for i in range(1, min):
  if (num1%i == 0) and (num2%i == 0):
    hcf = i
    print(i, end = " ")
  i = i + 1
print("\nHcf of the number is: ",hcf)

lcm_result = (num1 * num2) / hcf  #rules to find lcm is a * b = hcf *lcm. so just divide hcf with a and b. and the result will be lcm
print("LCM result is:", lcm_result)

```

27. Write a C program to convert Binary to Octal number system.

```python
#Binary - Decimal
#Decimal - Octal

import math
num = int(input("Enter a binary number"))
dec= 0
place = 0
last = 0

while num != 0:
  last = num % 10
  dec = dec + last * pow(2, place)
  num = num // 10
  place = place + 1
print("Decimal value is:",dec)

oct = 0
place = 1
last = 0

while dec != 0:
  last = dec % 8
  oct = oct + last * place
  dec = dec // 8
  place = place * 10
print("Octal value is:",oct)
```
